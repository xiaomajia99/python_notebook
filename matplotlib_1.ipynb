```python
from matplotlib import pyplot as plt
import numpy as np

x = [1,2,3,1]
y = [1,3,0,1]
plt.plot(x,y,color='r',linewidth='2',linestyle='--',marker='D',label='One')
plt.plot(np.array(x)+1,np.array(y)+1,color='g',linewidth='2',linestyle=':',marker='o',label='Two')
plt.plot(np.array(x)+2,np.array(y)+2,color='k',linewidth='2',linestyle='-.',marker='>',label='Three')

plt.title('GIF')
plt.ylabel('Y_LABEL')
plt.xlabel('X_LABEL')

#plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,6])

plt.ylim([0,10])
plt.xlim([0,10])

plt.grid(True)
plt.legend(loc="upper left")
plt.show()
```
